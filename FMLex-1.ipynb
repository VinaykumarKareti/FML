```python
X_train,X_test,y_train,y_test=train_test_split(df[["TV"]],d[["Sales"]],test_size=0.3,random_state=2)
```

```python
X_train,y_train
```

```python
model=linear_model.LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
mseloss=mean_squared_error(y_test,y_pred)
r2_loss2=r2_score(y_test,y_pred)
mae_loss=mean_absolute_error(y_test,y_pred)
print("MSE :",mseloss,"\nMAE :",mae_loss,"\nr2_score :",r2_loss2,"\nRMSE :",(mseloss)**(1/2))
```
